```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(3,4, 'ro', label = 'red circle')
plt.plot(2,5, 'bo', label = 'blue circle')
plt.plot(1,6, 'go', label = 'green circle')
plt.legend()
plt.show()
```

```python
x = [0,1,3]
y = [0,2, -1, ]
plt.plot(x,y, 'm')
plt.show()

```

```python
x = [0,3]
y = [3,0]
plt.plot(x, y)
plt.show()
```

```python
plt.plot(np.random.randn(10))
plt.plot(np.random.randn(10))
plt.show()
```
