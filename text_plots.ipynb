```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(3,4, 'ro', label = 'red circle')
plt.plot(2,5, 'bo', label = 'blue circle')
plt.plot(1,6, 'go', label = 'green circle')
plt.legend()
plt.show()
```

```python
x = [0,1,3]
y = [0,2, -1, ]
plt.plot(x,y, 'm')
plt.show()

```

```python
x = [0,3]
y = [3,0]
plt.plot(x, y)
plt.show()
```

```python
plt.plot(np.random.randn(10))
plt.plot(np.random.randn(10))
plt.show()
```

```python
#plt.subplot()
plt.subplot(2,2,1) # 2 rows, 2 colomns, 1st plot
plt.plot(np.random.randn(10))
plt.subplot(2,2,4) # 2 rows, 2 columns, 4th plot
plt.plot(np.random.randn(10))

```

```python
# subplots()
fig, ax = plt.subplots(2,2, figsize=(10,6)) # 2 rows with 2 columns
ax[0,0].plot(np.random.randn(10), 'r') # 1st plot 1st row
ax[0,0].plot(np.random.randn(10))
ax[1,0].plot(np.random.randn(4,4)) # 2nd row 1st plot

plt.show()
```

```python
# create a 3x3 subplot geometry
# user for loop .flatten()

fig, ax = plt.subplots(3,3)
plt.tight_layout()
for i, ax in enumerate(ax.flatten()):
    if i == 0:
        ax.plot(np.random.randn(2,2))
        ax.set_title('first')
    elif i == 1:
        ax.plot(np.random.randn(3,3))
        ax.set_title('second')

    else:
        ax.plot(np.random.randn(4,4))
plt.show()
```

```python
# Making graphs look nicer

x = np.linspace(-3,3,101)
plt.plot(x,x, label = 'y=x')
plt.plot(x,x**2, label = 'y=x^2')
plt.plot(x,x**3, label = 'y=x^3')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Plot')
plt.xlim([x[0],x[-1]])

plt.show()
```
